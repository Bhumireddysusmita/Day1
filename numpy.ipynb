```python
import numpy as np
empArr = np.empty((3, 4))
print(empArr)
```

```python
full_arr=np.full((3,4),5)
print(full_arr)
```

```python
arr=np.zeros(4)
print(arr)
```

```python
arr2=np.zeros((3,5),dtype=int)
print(arr2)
```

```python
arr=np.ones(4)
print(arr)
```

```python
arr1=np.ones((4,3),dtype=int)
print(arr1)
```

```python
arr=np.array([[1,4],[3,4]])
row=[3,4]
print(row in arr.tolist())
```

```python
arr=np.array([[1,2,3],[42,np.nan,np.nan]])
print(arr)
cleaned_arr=arr[~np.isnan(arr).any(axis=1)]
print(cleaned_arr)
```

```python
arr=np.array([[[2,2,2],[2,2,2]]])
print(arr.shape)
arr2=np.squeeze(arr)
print(arr2.shape)
```

```python
arr=np.array([[2, 8, 9, 4],
                   [9, 4, 9, 4],
                   [4, 5, 9, 7],
                   [2, 9, 4, 3]])
arr1=repr(arr).count("8, 9, 4")
print(arr1)
```

```python
x = np.array([1,2,3,4,5,1,2,1,1,1])
print(x)
print(np.bincount(x).argmax())
```

```python
num1 = np.array([1, 2])
num2 = np.array([[10, 20],
                 [30, 40]])

for a, b in np.nditer([num1,num2]):
  print(a, ":", b)
```

```python
a = np.array([1, 2])
b = np.array([4, 6])
res = np.array(np.meshgrid(a, b)).T.reshape(-1, 2)
print(res)
```
